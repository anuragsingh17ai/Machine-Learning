# Pearson Correlation - Bivariate and Multivariate Analysis (for numeric-numeric)

### Pearson Correlation Formula

The Pearson correlation coefficient, denoted as \( r \), is calculated using the following formula:

$$
r = \frac{\sum{(x_i - \overline{x})(y_i - \overline{y})}}{\sqrt{\sum{(x_i - \overline{x})^2} \sum{(y_i - \overline{y})^2}}}
$$

Where:
- \( x_i \) and \( y_i \) are the individual data points for variables \( x \) and \( y \)
- \( \overline{x} \) and \( \overline{y} \) are the means of the variables \( x \) and \( y \)
- \( \sum \) represents the sum over all data points


## **1. Pearson Correlation (Bivariate Analysis)**
- **Definition**: Pearson correlation measures the strength and direction of the linear relationship between two numerical variables.
- **Coefficient (r)**:
  - **+1**: Perfect positive linear relationship (both variables increase together).
  - **-1**: Perfect negative linear relationship (one variable increases, the other decreases).
  - **0**: No linear relationship between the variables.

## **2. Pearson Correlation (Multivariate Analysis)**
- **Definition**: In multivariate analysis, Pearson correlation can be used to examine pairwise relationships between multiple numerical variables.
- **Interpretation**: Each pair of variables will have its own Pearson correlation coefficient, helping to identify relationships within a dataset.

### **Usage in Data Analysis**
- **Identifying Patterns**: Use Pearson correlation to understand how variables are related, which can guide further analysis or model selection.
- **Caution**: Pearson correlation only captures linear relationships and can be sensitive to outliers.

### **Assumption for Corrrelation**
- **Continuous data**
- **Linear Relationship**
- **Homoskedastic relationship**: means that the errors (or mistakes) in our predictions are evenly spread out no matter what value we are looking at.





### Understanding P-Values in Simple Terms

A **p-value** is a statistical measure that helps you determine whether your data provides enough evidence to reject a null hypothesis.

- **Null Hypothesis (H₀)**: This is usually a statement that there's no effect or no relationship between variables. For example, in the context of correlation, the null hypothesis might be that two variables are not correlated (i.e., their Pearson correlation coefficient is zero).
  
- **P-Value**: The p-value tells you the probability of observing your data, or something more extreme, if the null hypothesis is true. In simpler terms, it's a measure of how likely your data could occur by random chance if there's actually no real effect or relationship.

  - **Low p-value (typically ≤ 0.05)**: Indicates that the observed data is unlikely under the null hypothesis, so you might reject the null hypothesis. This suggests that there is likely a significant effect or relationship.
  - **High p-value (> 0.05)**: Indicates that the data is not unusual under the null hypothesis, so you fail to reject the null hypothesis. This suggests that any observed effect could just be due to random chance.

### How P-Values are Related to the Pearson Correlation Coefficient

The **Pearson correlation coefficient (r)** measures the strength and direction of the linear relationship between two variables. But to determine if this relationship is statistically significant (i.e., not due to random chance), you need the **p-value** associated with the Pearson correlation.

- **P-Value in Correlation**: When you calculate the Pearson correlation, you often also get a p-value. This p-value tells you whether the correlation you observed is statistically significant.
  
  - **If the p-value is low (e.g., ≤ 0.05)**: You can reject the null hypothesis that there's no correlation (r = 0). This means there's likely a true relationship between the variables, and the Pearson correlation coefficient you calculated is significant.
  
  - **If the p-value is high (e.g., > 0.05)**: You fail to reject the null hypothesis. This means the correlation you observed might just be due to random chance, and there's no strong evidence of a real relationship between the variables.

### Example

Imagine you calculate the Pearson correlation coefficient between two variables (e.g., hours studied and test scores) and find \( r = 0.8 \), suggesting a strong positive relationship. But is this correlation statistically significant?

- **If the p-value is 0.01**: This means there's only a 1% chance that such a strong correlation could occur if there's no real relationship between the variables. Since this is very unlikely, you reject the null hypothesis and conclude that the correlation is significant.
  
- **If the p-value is 0.3**: This means there's a 30% chance that the observed correlation could occur by random chance, which is quite high. So, you fail to reject the null hypothesis, suggesting that the correlation might not be significant.

### Summary

- **Pearson Correlation Coefficient (r)**: Measures the strength and direction of a linear relationship between two variables.
- **P-Value**: Helps you determine whether the observed correlation is statistically significant or could have happened by random chance.
- **Low P-Value**: Suggests a significant correlation.
- **High P-Value**: Suggests the correlation might not be significant and could be due to random variation.


```python
import numpy as np

height  = [60, 62, 65, 68, 70, 74]
weight = [140, 138, 150, 166, 190, 250]

np.corrcoef(height, weight)

# Output:
#             heigth     weight
#    height [[1.        , 0.92989745],
#    weight [0.92989745, 1.        ]])

```

```python
import pandas as pd

df = pd.read_csv("https://www.ishelp.info/data/insurance.csv")
df.head()
```

```python
df[['age','bmi','children','charges']].corr()  # same as np.corrcoef
```

```python
df['charges'].corr(df['bmi'])
```

```python
from scipy import stats

corr = stats.pearsonr(df['charges'], df['age'])
corr #lesser the pvalue better it is
```

```python
round(corr[1], 30)
```

```python
corr_df = pd.DataFrame(columns=['r','p'])
for col in df:
  if pd.api.types.is_numeric_dtype(df[col]):
      r, p = stats.pearsonr(df.charges, df[col])
      corr_df.loc[col]=[round(r,3), round(p,3)]
```
