```python
import pandas as pd
import seaborn as sns
df = pd.read_csv("/content/tested.csv")
```

### 1. Categorical Data

#### a. Countplot

```python
sns.countplot(x=df['Survived'])
# df['Survived'].value_counts().plot(kind="bar")
```

```python
sns.countplot(x=df['Pclass'])
```

```python
mean_of_pclass_1 = df[df['Pclass']==1]['Fare'].mean()
mean_of_pclass_2 = df[df['Pclass']==2]['Fare'].mean()
mean_of_pclass_3 = df[df['Pclass']==3]['Fare'].mean()

mean_of_pclass_1, mean_of_pclass_2, mean_of_pclass_3
```

```python
sns.countplot(df["Embarked"])
```

#### b. PieChart

```python
df["Survived"].value_counts().plot(kind="pie", autopct ="%.2f")
```

```python
df["Sex"].value_counts().plot(kind="pie", autopct ="%.2f")
```

### 2. Numerical_data

#### a. Histogram


```python
import matplotlib.pyplot as plt
plt.hist(df["Age"]) # to understand distribution of data in numerical column
```

#### b. Distplot


```python
sns.distplot(x=df['Age']) # probabilit density function
# probability of age 40 is 0.015
```

#### c. Boxplot

* "Minimum" = Q1 - 1.5 * IQR
* Q1 = 25 percentile
* Q3 = 75 percentile
* IQR = Q1 - Q3
* "Maximum" = Q3 + 1.5 * IQR

* below Minimum and above Maximum are potential Outliers

```python
sns.boxplot(x = df["Fare"])
```

```python
sns.boxplot(x = df["Age"])
```
