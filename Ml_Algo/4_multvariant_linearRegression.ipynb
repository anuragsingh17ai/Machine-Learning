```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from word2number import w2n
import math
```

```python
df=pd.read_csv('csv_files/hiring.csv')
df
```

```python
df.experience=df.experience.fillna('zero')
df
```

```python
df['test_score(out of 10)']=df['test_score(out of 10)'].fillna(math.floor(df['test_score(out of 10)'].median()))
df
```

```python
df.experience = df.experience.apply(w2n.word_to_num)
df
```

```python
regr=LinearRegression()
regr.fit(df[['experience','test_score(out of 10)','interview_score(out of 10)']],df['salary($)'])
regr.predict([[2,9,6]])
```
