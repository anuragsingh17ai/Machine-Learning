```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df=pd.read_csv('csv_files/canada_per_capita_income.csv')
df
```

```python
regr=linear_model.LinearRegression()
```

```python
plt.xlabel('year')
plt.ylabel('income in US ($)')
plt.title('per capita income of canada')
plt.scatter(df.year,df['per capita income (US$)'],c='red',marker='+')
```

```python
x=np.array(df.year)
x=x.reshape(-1,1)

regr.fit(x,df['per capita income (US$)'])
predicted_price=regr.predict(x)
predicted_price
```

```python
plt.xlabel('year')
plt.ylabel('income in US ($)')
plt.title('per capita income of canada')
plt.scatter(df.year,df['per capita income (US$)'],c='red',marker='+')
plt.plot(x,predicted_price,c='blue')
```
